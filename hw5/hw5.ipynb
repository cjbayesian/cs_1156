```python
def gradient_step(u,v,eta=0.1):
    grd_u = (u*exp(v) - 2*v*exp(-u)) * (4*v*exp(-u) + 2*exp(v))
    grd_v = (u*exp(v) - 2*v*exp(-u)) * (2*u*exp(v) - 4*exp(-u))
    new_u = u - (eta*grd_u)
    new_v = v - (eta*grd_v)
    return (new_u,new_v)
    
def Error(u,v):
    return (u*exp(v) - 2 * v * exp(-u) )**2

u,v = 1.0,1.0
e = 1
i = 0
while e > 10e-14:
    i = i + 1
    u,v = gradient_step(u,v)
    e = Error(u,v)
    print i,u,v,e
```

### Question 7
Now step-wise, or as I like to call it: etch-a-sketch style.

```python
def etch_a_sketch(u,v,u_step=True,eta=0.1):
    if u_step:
        grd_u = (u*exp(v) - 2*v*exp(-u)) * (4*v*exp(-u) + 2*exp(v))
        new = u - (eta*grd_u)
    else:
        grd_v = (u*exp(v) - 2*v*exp(-u)) * (2*u*exp(v) - 4*exp(-u))
        new = v - (eta*grd_v)
    return new

u,v = 1.0,1.0
e = 1
i = 0
for rep in range(150):
    i = i + 1
    u = etch_a_sketch(u,v,u_step=True)
    e = Error(u,v)
    print i,u,v,e
    
    i = i + 1
    v = etch_a_sketch(u,v,u_step=False)
    e = Error(u,v)
    print i,u,v,e 
```

## Logistic Regression



```python
def logistic_gradient(x,y,w):
    grd = y*x / ( 1 + exp(y * np.dot(x,w)))
    return grd
```

```python
def lineFromPoints(p1,p2):
    '''Return slope and intercept of a line defined by two points '''
    # y = m x + b
    m = (p1[1] - p2[1]) / (p1[0] - p2[0])
    b = p1[1] - m * p1[0]
    return (m,b)
    
def classifyByLine(x,m,b):
    '''Return true is below the line, false if above'''
    y = x[:,1] > x[:,0] * m + b
    return y

def clsfn2(x,w):
    '''classify by a random line'''
    m,b = lineFromPoints(np.random.rand(2)*2-1,np.random.rand(2)*2-1)
    y = classifyByLine(x,m,b)
    return y

d = simData(N=1000,classifyfn=clsfn2,w=None)
plot_perc(d)
```
