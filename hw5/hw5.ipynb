### Question 2
Nonlinear Transforms
$$\Phi(1, x_1, x_2) = (1, x_1^2, x_2^2)$$

```python
def simData(N,classifyfn,w):
    ''' Simulate N points on R2 [-1,1]'''
    x = np.random.rand(N,2)*2-1
    
    # pick 2 points to draw a line through
    sp = np.random.rand(2,2)*2-1
    
    y = classifyfn(x,w)
    
    return {'x':x,'y':y}

def transform(x):
    x = x**2
    return x

def clsfn(x,w):
    '''perceptron, implementing h(x)=sign(W^T . x)'''
    #s = w[0] + w[1] * x[:,0] + w[2] * x[:,1] 
    # Let's use matrix notation instead to allow
    # generalization to n-dimensional space
    x = transform(x)
    x = np.column_stack((np.ones(x.shape[0]).T,x))
    s = (w * x).sum(1)
    y = (s > 0)
    return y

def plot_perc(data,pdf_w=None):
    xx = np.array([-1,1])
    y_color = np.array(['red' if i==1 else 'blue' for i in data['y']])

    plt.scatter(data['x'][:,0],data['x'][:,1],c=y_color,s=100)
    plt.axis([-1, 1, -1, 1])
    
    ## If given a pdf object then save to page
    if pdf_w:
        pdf_w.savefig()  # saves the current figure into a pdf page
        plt.close()    
    ## Else, show the plot inline
    else:
        plt.show()
```

```python
# Somehow, my classification scheme is inverted.
d = simData(N=1000,classifyfn=clsfn,w=np.array([-1,4,-4]))
plot_perc(d)
```
