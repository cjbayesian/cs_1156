## Linear Regression

Next let's implement linear regression.
The closed form solution to minimizing the squared error can be written in matrix form:

$\mathbf{W} = (\mathbf{X}^T\mathbf{X})^{-1} \mathbf{X}^T \mathbf{y}$

```python
def linear_reg_fit(x,y):
    '''Let's get our linear algebra on.
    Numpy can help!'''
    inv = np.linalg.inv # for inverting matrices
    return np.dot(np.dot(inv(np.dot(x.T,x)),x.T),y)
```

We'll also need our utility functions from hw1:

```python
def lineFromW(x,w):
    '''Return y for a given x and a set of weights w. For drawing the perceptron'''
    # w0 + w1 x + w2 y = 0
    # y = (-w0 - w1 x) / w2
    y = (-w[0] - w[1] * x) / w[2]
    return y

def lineFromPoints(p1,p2):
    '''Return slope and intercept of a line defined by two points '''
    # y = m x + b
    m = (p1[1] - p2[1]) / (p1[0] - p2[0])
    b = p1[1] - m * p1[0]
    return (m,b)

def lineFromMB(x,m,b):
    '''Return y for a given x, m=slope, and b=y intercept'''
    y = m * x + b
    return y
    
def classifyByLine(x,m,b):
    '''Return true is below the line, false if above'''
    y = x[:,1] > x[:,0] * m + b
    return y

def classifyByW(x,w):
    '''Linear perceptron, implementing h(x)=sign(W^T . x)'''
    #s = w[0] + w[1] * x[:,0] + w[2] * x[:,1] 
    # Let's use matrix notation instead to allow
    # generalization to n-dimensional space
    x = np.column_stack((np.ones(x.shape[0]).T,x))
    s = (w * x).sum(1)
    y = (s > 0)
    return y

def simData(N=20):
    ''' Simulate N points separated by a random line '''
    x = np.random.rand(N,2)*2-1
    
    # pick 2 points to draw a line through
    sp = np.random.rand(2,2)*2-1
    m,b = lineFromPoints(sp[0,:],sp[1,:])
    
    # define y according to that line
    y = classifyByLine(x=x,m=m,b=b)
        
    return {'x':x,'y':y}
```

And a funtion to visualize:

```python
# Function to plot each page of a multipage pdf to visualize the PLA
def plot_perc(data,w,pdf_w=None):
    xx = np.array([-1,1])
    y_color = np.array(['red' if i==1 else 'blue' for i in data['y']])

    plt.scatter(data['x'][:,0],data['x'][:,1],c=y_color,s=100)
    plt.axis([-1, 1, -1, 1])
    #plt.plot(xx,lineFromMB(xx,m,b))

    plt.plot(xx,lineFromW(xx,w))
    y_perceptron = classifyByW(data['x'],w)
    correct = np.logical_not(np.logical_xor(y_perceptron,data['y']))

    y_perceptron = ['*' if i else 'x' for i in correct]

    for i in range(data['x'].shape[0]):
        plt.scatter(data['x'][i,0],data['x'][i,1],marker=y_perceptron[i],color='white',s=100)
        
    ## If given a pdf object then save to page
    if pdf_w:
        pdf_w.savefig()  # saves the current figure into a pdf page
        plt.close()    
    ## Else, show the plot inline
    else:
        plt.show()
```

Next lets define a linear regression experiment where we sim some data, fit a linear regression and evaluate $E_{in}$.

```python
def lr_experiment(N=100,plotres=True,sim_fnc=simData):
    # Simulate some random, linearly separable data
    data = sim_fnc(N=N)

    # Add on the x0 = 1 artificial coordinate
    x = data['x']
    x = np.column_stack((np.ones(x.shape[0]).T,x))

    # encode y as {-1,1} rather than {true,false}
    y = data['y']
    y = np.array([1 if yy else -1 for yy in y])

    # Fit a linear regression to the data
    w = linear_reg_fit(x,y)
    
    # Predict y from the fitted regression
    y_linreg = classifyByW(data['x'],w)

    # Evaluate E_in
    E_in = np.logical_xor(y_linreg,data['y']).mean()
    
    if plotres:
        # Plot the results
        plot_perc(data=data,w=w)
    
    return E_in
```

## Nonlinear Transformation
Remembering that linear regression requires only _linearity in weights_, we can do non-linear transforms on $x$ and still apply linear regression. Here we will use the transformation function:

$$f(x_1,x_2) = sign(x_1^2 + x_2^2 - 0.6) $$

First, we'll set up a function to simulate data from $f$ with added noise.

```python
def sim_nonlinear(N=10,noise=0.1,plotsim=False):
    ''' Simulate N from sign(x_1^2 + x_2^2 - 0.6) + noise '''
    x = np.random.rand(N,2)*2-1
    
    y = ( np.power(x[:,0],2) + np.power(x[:,1],2) - 0.6 ) > 0
    
    # Choose some random y values to flip (eg adding noise)
    flip_idx = random.choice(N,int(N*noise))
    y[flip_idx] = np.logical_not(y[flip_idx])
    
    return {'x':x,'y':y}

def plotsim(data):
    '''plot simulated data'''
    xx = np.array([-1,1])
    y_color = np.array(['red' if i==1 else 'blue' for i in data['y']])

    plt.scatter(data['x'][:,0],data['x'][:,1],c=y_color,s=100)
    plt.axis([-1, 1, -1, 1])
    plt.show()
```

### Question 10
What is $E_{out}$ for our new hypothesis?

```python
def lr_experiment3(N_in=100,N_out=1000,sim_fnc=sim_nonlinear):
    # Simulate some random, linearly separable data
    N = N_in + N_out
    data = sim_fnc(N=N)

    # Transform to (1,x_1,x_2,x_1x_2,x_1^2,x_2^2)
    x = data['x'][:N_in]
    x = np.column_stack((np.ones(x.shape[0]).T,x,x[:,0]*x[:,1],x[:,0]**2,x[:,1]**2))

    # encode y as {-1,1} rather than {true,false}
    y = data['y'][:N_in]
    y = np.array([1 if yy else -1 for yy in y])

    # Fit a linear regression to the data
    w = linear_reg_fit(x,y)
    
    # Predict y_out from the fitted regression
    x = data['x'][N_in+1:]
    x = np.column_stack((x,x[:,0]*x[:,1],x[:,0]**2,x[:,1]**2))
    y_linreg = classifyByW(x,w)

    # Evaluate E_out
    E_out = np.logical_xor(y_linreg,data['y'][N_in+1:]).mean()
    
    return E_out
```

```python
## Run a few times:
for r in range(5):
    print lr_experiment3()
```

Our $E_{out}$  is about equal to the amount on noise we injected into the simulated data. Using our non-linear transformation of the feature space gives us the best performance we could hope for. The remaining 10% is not encoded in $x$ in any way so no matter what we do, we can't get better than this. 

This is a great example of why even when you have **big data**, there is a limit on how well your model can perform. At a maximum, you can only do as well as the degree to which $y$ is encoded by $x$. 

Let's do one final experiment to demonstrate this. Instead of having a sample size of 1000, we have a **big data**<sup>tm</sup> training set of 1 million datapoints. We're going to rock the socks off of those punks their puny $n=1000$ dataset.  

```python
lr_experiment3(N_in=1000000,N_out=1000)
```

There is no sample size, magical algorithmic cleverness, or feature engineering which is going to change the fact that 10% of the variation in $y$ is _not signal_.
